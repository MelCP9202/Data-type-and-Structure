## Theory Base Questions

1. What are data structures, and why are they important?
- Data Structures are specialised formats or containers used to store and organise data efficiently. They provide various ways to store, access, and manipulate data according to specific needs, which is why they are important.

2. Explain the difference between mutable and immutable data types with example.
- Mutable means you can change the contents of an object after it is created. List is one of the example of a mutable object as we can perform various operations like adding, removing or sorting the elements. Immutable on the other hand, means you can't change the contents after it is created. This can be useful for situations when you want to ensure that the data remains constant. Tuple is an example of an immutable type.

3. What are the main differences between lists and tuples in Python?
- The main differences between lists and tuples are: list is mutable while tuple is immutable, lists have several built in methods while tuples have fewer. List consumes more memory and tuples consumes less memory.

4. Describe how dictionaries store data.
- Dictionaries are key-value pairs that allows to store and retrieve data base on a unique key. The keys must be immutable, such as strings or numbers.

5. Why might you use a set instead of a list in Python?
- Sets are unordered list of unique elements, and can be used when needed to work with distinct items. It does not allow duplicate values, which is why we use a set instead of a list in Python.

6. what is a string in Python, and how is it different from a list?
- A string is a sequence of characters between single or double quotation, while a list is a collection of ordered items or a sequence of items, where each item could be anything (an integer, float, a string, etc.). Both of them have lengths; a string's length is the number of characters in the string, a list's length is the number of items in the list.

7. How do tuples ensure dats integrity in Python?
- Tuples are immutable which prevents data modification as a result it ensures data integrity in Python.

8. What is a hash table, and how does it relate to dictionaries in Python?
- Hash table is defined as a data structure used to insert, look-up, and remove key-value pairs quickly. In hash table each key is translated by hash function into a distinct index in an array which means it maps the key with the value and we know that dictionaries are a key-value pair. Therefore, we can say dictionaries in Python are an implemantion of hash table.

9.  Can list contain different data types in Python?
- No, a list can't contain different data types in Python.

10. Explain why strings are immutable in Python.
- Strings are immutable in Python,because immutability makes it hashable (usuable as dictionary), memory efficient, and thread safe, ensuring they can be reused without any unexpected changes.

11. What are the advantages do dictionaries offer over lists for certain tasks?
- Dictionaries offer quick access to data. Instead of looking at a list one by one, the key can be used to instantly find what we are looking for.

12. Describe a scenario where using a tuple would be preferred over a list.
- Tuples are preferred when we need to use an oredered sequence which need no changes, i.e, tuples are used for immutable items/objects. Imagine a scenario where we want to list the name of students who got full marks in math, english, and science in a particular year. We can represent it via a tuple.

13. How do sets handle duplicate in Pyhton?
- Duplicate elements are not allowed in sets. If tried to insert the same item again, it overwrites previous one or ignore the duplicate item.

14. How does the "in" keyword work differently for list and dictionaries?
- The "in" keyword is a membership operator and works differently for lists and dictionaries. In a list, when "in" is used, it iterates through each element of the list, comparing each value to the searched value. In short, it checks if the specific value is present in the list or not. Meanwhile, in a dictionary, it is used to check the presence of a key but not the value, i.e., it checks if the specified value exists as a key within the dictionary.

15. Can you modify the elements of a tuple? Explain why or why not.
- No, we can't modify the elements of a tuple, because tuples are immutable.

16. What is a nested dictionary, and give an example of its use case?
- Dictionaries can contain other dictionaries as values, forming a nested structure, is called a nested dictionary. It helps organize complex or grouped data in a clean and structure way. For example; student data: each student having attribute like name, age, grade, course enrolled, etc.

17. Describe the time complexity of accessing elements in a dictionary.
- The time complexity of accessing elements or a value by its key in a dictionary is O(1). This is because dictionaries use a hash table internally which allows constant time lookup and hence it is fast and an efficient way to store and retrieve data using key-value pair.

18. In what situations are list preferred over dictionaries?
- When we need an oredered collection of items or an ordered sequence of data, lists are preferred over dictionaries.

19. Why are dictionaries considered unordered, and how does that affect data retrieval?
- A dictionary contains key-value pairs but does not possess an order for the pairs as a result they don't maintain the order of elements inserted, and data is not stored squentially. Because it is unordered we cannot retrieve data by indexing (by their position), instead we have to use its unique key, which provides faster access to a value.

20. Explain the difference between a list and a dictionary in terms of data retrieval.
- In a list data is retrieved via indices while in a dictionary it is via key.
